````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
````

````python
def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory

inventory = initialize_inventory(products)
````

````python
def get_customer_orders():
    orders_number = int(input("Enter the number of customer orders: "))

    customer_orders = { input("Enter the name of a product that a customer wants to order: ") for i in range(orders_number) }

    return customer_orders


customer_orders = get_customer_orders()

    

    
````

````python
total_products_ordered = len(customer_orders)
percentage_of_products_ordered = (total_products_ordered / len(products))*100


print("Order Statistics")
print("Total Products Ordered: ", total_products_ordered)
print("Percentage of Unique Products Ordered: ", percentage_of_products_ordered) 
````

````python
#Add a new function to calculate the total price of the customer order. 
#For each product in customer_orders, prompt the user to enter the price of that product.
#Use comprehension to calculate the total price. Note: assume that the user can only have 1 unit of each product.

def total_order_price(customer_orders):

    prices = [ float(input(f"Enter the price of {product} :")) for product in customer_orders]

    total = sum(prices)

    return total 

prices = total_order_price(customer_orders)
print("Total price : ", prices)
````

````python
#Modify the update_inventory function to remove the product from the inventory if its quantity becomes zero 
#after fulfilling the customer orders.
#Use comprehension to filter out the products with a quantity of zero from the inventory.

def update_inventory(customer_orders, inventory):

    updated_inventory = { product : (inventory[product] -1 if product in customer_orders else inventory[product]) for product in inventory}
    updated_inventory_0 = { product : quantity for product, quantity in updated_inventory.items() if quantity > 0}

    return updated_inventory_0
    
inventory = update_inventory(customer_orders, inventory)
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")
````
